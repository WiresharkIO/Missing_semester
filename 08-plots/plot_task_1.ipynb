```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = 'D:\Course_Content\Missing_semester\Exercise_8\data.xlsx'
```

```python
df_data = pd.read_excel(data, header=None)
# df_data
player_one_scores = df_data.iloc[:, 0]
player_two_scores = df_data.iloc[:, 1]
```

```python
df_data
```

```python
# a.1
plt.xlim(0.1, 100)
plt.ylim(0.1, 100)

# a.2
plt.xscale('log')
plt.yscale('log')

# a.3, a.6
plt.scatter(player_one_scores, player_two_scores, marker='^', color='blue', label='Results')

# a.4
plt.grid(True, which='both', linestyle='--', linewidth=0.4)

# a.5, a.8
plt.xlabel('Player One', fontsize=20, color='#FF0000')
plt.ylabel('Player Two', fontsize=20, color='#FF0000')

# a.7, a.8
# reference - https://stackoverflow.com/questions/23698850/manually-set-color-of-points-in-legend
legend=plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=1, fontsize=20)
legend.get_frame().set_color('#FF0000')

plt.show()
```
