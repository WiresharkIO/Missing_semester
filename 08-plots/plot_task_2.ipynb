```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = 'D:\Course_Content\Missing_semester\Exercise_8\data.xlsx'
```

```python
df_data = pd.read_excel(data, header=None)
player_one_scores = df_data.iloc[:, 0]
player_two_scores = df_data.iloc[:, 1]
```

```python
plt.xlim(0.1, 100)
plt.ylim(0.1, 100)
plt.xscale('log')
plt.yscale('log')
plt.scatter(player_one_scores, player_two_scores, marker='^', color='blue', label='Results')
plt.grid(True, which='both', linestyle='--', linewidth=0.4)
plt.xlabel('Player One', fontsize=20, color='#FF0000')
plt.ylabel('Player Two', fontsize=20, color='#FF0000')
legend=plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=1, fontsize=20)
legend.get_frame().set_color('#FF0000')

# until here everything is similar to task_1

# b.1 to plot a diagonal such that on the diagonal player1 score is same as player2 score..
# Top of diagonal-line means player2 score is higher, bottom of diagonal means player1 score is higher..
plt.plot([0.1, 100], [0.1, 100], linestyle='--', color='blue')

# b.2 to plot diagonal to divide score higher than (player1 score * 10) and (player2 score), so the limit 
# of x-axis which represents player1 score starts 1 and corresponding start of player2 score is @ 0.1 and 
# this continues for further limits across the range in both the axes.
plt.plot([0.1, 100], [0.01, 10], linestyle='--', color='black')
# based on trial and error, this label for diagonal was created
plt.text(5.7, 0.75, '[Player1 * 10]', fontsize=10, rotation=37, color='black')

# b.3 same as previous, but here point of interest is player2 score..
plt.plot([0.01, 10], [0.1, 100], linestyle='--', color='grey')
# based on trial and error, this label for diagonal was created
plt.text(0.45, 6, '[Player2 * 10]', fontsize=10, rotation=36, color='grey')

plt.show()
```
